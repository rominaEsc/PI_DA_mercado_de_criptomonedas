```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns

# Establecer la opción para mostrar todas las columnas
pd.set_option('display.max_columns', None)

# No mostrar advertencias
pd.options.mode.chained_assignment = None
```

La capitalización de mercado(Capitalización de Mercado = Precio Actual × Suministro en Circulación) es una métrica importante porque proporciona una idea general del tamaño relativo de una criptomoneda en comparación con otras y del valor total de un mercado en particular. Las criptomonedas con una alta capitalización de mercado suelen ser ampliamente conocidas y tienen una gran influencia en el mercado en general. Por esto, buscaremos conocer las 10 cryptomonedas con mayor capitalizacion de mercado.

Procederemos a cargar el csv que obtuvimos de la api.

```python
cryptos_cap_mercado_top10_df = pd.read_csv('data/cryptos_cap_mercado_top10_df.csv')
```

```python
cryptos_cap_mercado_top10_df
```

Las 10 monedas con mayor capitalización de mercado segun la api de coingecko son:

```python
cryptos_cap_mercado_top10_df['id']
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

sns.barplot(y=cryptos_cap_mercado_top10_df['id'], x= cryptos_cap_mercado_top10_df['capitalizacion_de_mercado'])
plt.title("Top 10 - Capitalizaciones de mercado")
ax.set_ylabel('Criptomoneda')
ax.set_xlabel('Capitalizacion de mercado')
plt.show()
```

En el gráfico anterior podemos notar que la capitalización de bitcoin es muy superior a la de las otras cryptomonedas analizadas. Pero para mostrarlo mejor lo pasaremos a porcentajes.

```python
aux1 = cryptos_cap_mercado_top10_df[['id', 'capitalizacion_de_mercado']]
bitcoin_market_cap = aux1[aux1['id'] == 'bitcoin']['capitalizacion_de_mercado'].values[0]
aux1['%_con_respecto_a_bitcoin'] = aux1['capitalizacion_de_mercado'] / bitcoin_market_cap * 100

aux1
```

```python
aux1.iloc[1:10].sum()
```

En la tabla anterior, podemos ver que la capitalización de mercado de bitcoin es mayor que el de las 9 cryptomonedas restantes y que estas suman el 80% de la capitalización de mercado de bitcoin. También notamos que Ethereum representa la mitad de aquel valor. Esto nos muestra que la capitalización de bitcoin es significativamente superior a la de las otras criptomonedas. Sin embargo, cabe destacar que el ecosistema crypto se esta diversificando, dando lugar a nuevos proyectos, tal como lo demuestran estos porcentajes que acabamos de mencionar.

Antes de continuar con el análisis definiremos los siguientes términos.

- id --> Identificador único de la criptomoneda en una plataforma o base de datos.
- precio_actual                       -->     Precio actual de la criptomoneda en una moneda de referencia (en este caso es en USD).
- capitalizacion_de_mercado	                        -->     Capitalización de mercado total de la criptomoneda, que es el precio actual multiplicado por la oferta circulante.
- valoración_totalmente_diluida             -->     Valoración total considerando todas las unidades que podrían existir si se minaran todas las monedas posibles. Se obtiene multiplicando el precio de mercado actual de un criptoactivo por su oferta máxima.
- volumen_total	                    -->     Volumen total de transacciones de la criptomoneda en un período de 24 horas.

```python
cryptos_cap_mercado_top10_df[['id','precio_actual','capitalizacion_de_mercado']].sort_values('precio_actual').reset_index(drop=True)
```

Podemos ver que las cryptomonedas con mayor precio_actual son Bitcoin, Ethereum y staked-ether superando 1600 USD. Mientras que el valor de dogecoin no llega ni a 0,1 USD

```python
aux = cryptos_cap_mercado_top10_df[['id','precio_actual','capitalizacion_de_mercado','valoración_totalmente_diluida','volumen_total']].sort_values('volumen_total').reset_index(drop=True)
aux
```

El volumen total de transacciones puede proporcionar información sobre la actividad comercial, la liquidez, las tendencias emergentes y la posible manipulación del mercado. De la tabla anterior podemos notar que la cryptomoneda que tuvo mayor cantidad de transacciones fue Tether, seguida de Bitcoin, Ethereum y usd-coin. 

Para mas información sobre la valoración totalmente diluida puede leer: 
* https://www.coingecko.com/learn/what-is-fully-diluted-valuation-fdv-in-crypto

Vamos a analizar la variación en el precio de las cryptomonedas en el día de la toma de datos. Pero antes de comenzar, consideramos que es bueno aclarar el contenido de las columnas a analizar.

- alto_24h	                        -->     Precio más alto alcanzado por la criptomoneda en las últimas 24 horas.
- bajo_24h	                            -->     Precio más bajo alcanzado por la criptomoneda en las últimas 24 horas.
- cambio_precio_24h                    -->     Cambio en el precio de la criptomoneda en las últimas 24 horas.
- cambio_precio_porcentual_24h	        -->     Porcentaje de cambio en el precio de la criptomoneda en las últimas 24 horas.
- cambio_en_cap_mercado_24h               -->     Cambio en la capitalización de mercado de la criptomoneda en las últimas 24 horas.
- cambio_en_cap_mercado_porcentual_24h    -->     Porcentaje de cambio en la capitalización de mercado de la criptomoneda en las últimas 24 horas.


```python
diario1 = cryptos_cap_mercado_top10_df[['id','precio_actual','alto_24h', 'bajo_24h', 'cambio_precio_24h', 'cambio_precio_porcentual_24h', 'cambio_en_cap_mercado_24h','cambio_en_cap_mercado_porcentual_24h' ]]
diario1
```

Podemos ver el porcentaje de cambio en el precio de las cryptomonedas en el día en que se realizó la toma de datos no supera el 7 %. Siendo Solana el cryptoactivo que presentó mayor variación. También podemos ver que Tether y usd_coin presentan valores de variación de precio por debajo del 0.1%, esto se debe a que al estar respaldadas por monedas fiducitarias, se las considera cryptomonedas estables (stablecoins) y su valor va a estar siempre cercano a 1 USD.  

Demanera muy similar al análisis anterior, podemos notar que el cambio porcentual de la capitalización de mercado de las cryptomonedas analizadas ha variado menos de un 7%, siendo Solana la cryptodivisa que presenta mayor variación. La capitalización de mercado se calcula en base a la siguiente fórmula: Capitalización de Mercado = Precio Actual × Suministro en Circulación. Por lo que las variaciones en el precio de un criptoactivo impactaran en su capitalizació de mercado.

Analizaremos ahora las columnas 'suministro_circulante','suministro_total','suministro_circulante_maximo', pero antes explicaremos estos términos:

- suministro_circulante	                -->     Cantidad de unidades de la criptomoneda en circulación y disponibles para ser negociadas.
- suministro_total	                    -->     Cantidad total de unidades de la criptomoneda que han sido generadas o minadas.
- suministro_circulante_maximo	        -->     Cantidad máxima de unidades de la criptomoneda que podrán existir en total, si está limitada.

```python
suministro = cryptos_cap_mercado_top10_df[['id','suministro_circulante','suministro_total','suministro_circulante_maximo']]
suministro
```

Podemos observar que existen cryptomonedas que poseen una cantidad máxima de 'suministro circulante' (unidades que podrán existir en total), como es el caso de bitcoin, binancecoin, ripple, staked-ether y cardano. 
Mientras que en otras como ethereum, solana y dogecoin no tienen este limite fijo. 
Otras como Tether y usd-coin están diseñada para mantener su valor en relación con una moneda fiduciaria, se mantienen respaldadas por dinero en cuentas bancarias, por lo que tampoco tienen este límite fijado. 
También podemos observar que para algunas cryptomonedas no todas las unidades totales están dispponibles para ser negociadas. Como son los casos de Solana y Cardano.

Continuaremos el análisis con las columnas 'ath', 'ath_cambio_porcentual', 'ath_fecha', 'atl', 'atl_cambio_porcentual', 'atl_fecha', pero antes explicaremos estos términos:
